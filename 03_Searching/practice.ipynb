```python
def binary_search(arr,target):
    start=0
    end=len(arr)-1
    while(start<=end):
        mid=(start+end)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]>target:
            end=mid-1
        elif arr[mid]<target:
            start=mid+1
    return -1
        
```

```python
arr=[3,5,8,12,15,21]
target=12
binary_search(arr,target)
```

```python
tar=7
binary_search(arr,tar)
```
