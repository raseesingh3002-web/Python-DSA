```python
def linear_search(array,target):
    size=len(array)
    for index in range(0,size):
        if(array[index]==target):
            return index
    return -1
```

```python
my_list=[10,23,45,70,11]
target=70

result =linear_search(my_list,target)
result
```

```python
target1=99
result1=linear_search(my_list,target1)
```
