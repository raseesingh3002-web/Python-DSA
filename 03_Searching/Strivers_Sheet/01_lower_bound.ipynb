### Lower Bound
Given a sorted array nums and an integer x, find the lower bound of x — the first (smallest) index where nums[index] >= x.  
If no such index exists (every element is smaller than x), return len(nums).  

Example 1  
Input: nums = [1,2,2,3], x = 2  

Output: 1  
(Index 1 is the smallest index where nums[i] >= 2.)  





Example 2  
Input: nums = [3,5,8,15,19], x = 9  

Output: 3  
(Index 3 is the smallest index where nums[i] >= 9.)

```python
def search(arr,x):
    start=0
    end=len(arr)-1
    ans=len(arr)
    while(start<=end):
        mid=(start+end)//2
        if arr[mid]>=x:
            ans=mid
            end=mid-1
        elif arr[mid]<x:
            start=mid+1
    return ans
```

```python
arr=[1,2,2,3]
x=2
search(arr,x)
```

```python
arr=[3,5,8,15,19]
x=9
search(arr,x)
```
