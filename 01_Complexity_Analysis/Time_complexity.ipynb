# TIME COMPLEXITY

Suppose there is a problem and there are two solutions of that problem one takes 1 second and another takes 20 minutes on the same input, but why?. This notebook that "why".

### 1. What is Time Complexity?

Time complexity measures how the amount of work an algorithm does increases as its input increases.

It does not measure time in seconds. Instead, it answers a different question: if I double my input, does my work double? Quadruple? Stay the same? That pattern of growth is the time complexity.

**Example:** Imagine a store receives 1000 orders, and we want to check if any order ID is duplicated. There are several ways to solve this problem, and two of them are as follows:

- One method compares every order with every other order, resulting in about 1000000 comparisons.  

- Another method goes through the orders once, keeping a list of IDs already seen. For each new order, it simply checks, "Is this ID on my list?" That requires about 1000 checks.

Both methods solve the same problem with the same input and give the same answer, but one does a thousand times more work. Now, consider 100000 orders: the first method does 10000000000 comparisons, while the second does only 100000. The gap keeps growing as the input increases.

Time complexity helps us see this difference before writing or running any code.

### 2. The Two Questions Anyone Writing Code Should Ask  

After writing code  there are two questions that an individual should ask:  

**1.) What is the time complexity of code?**  
Time complexity of code helps in answering "How much more work will code do as the input gets bigger?"  

**2.) Can I improve time complexity of code?**  
This answers you "can the same problem be solved in 1000 steps instead of 1000000 steps?"


### 3. Why Not Just Time the Code?

There is one answer of "how to measure the time that a code taken?" is that run it and measure the time with a stopwatch. This is called the experimental approach, and it fails due to several reasons:

1. **Different runs give different times.**  
Many times it happens that a lot of different processes are running in the background of the machine, which results in different timings when the same code is run more than once on the same machine.

2. **Small changes shift the numbers.** A different laptop or a
   different Python version changes the timing, even though the
   algorithm did not change.  

3. **Numbers, but no formula.** If 1,000 items took 2 seconds, we
   still cannot predict what 50,000 items will take. There is no
   relationship between input size and time.  

4. **Every solution must be fully implemented.** To compare three
   ideas by timing, all three must be coded and run — only to throw
   two away. Analysis on paper could rank them beforehand.  
   
5. **Large inputs take too long to test.** The inputs where speed
   matters most are exactly the ones where a single test run takes
   hours.

Notice the common thread: timing tells us about *this run, on this
machine, with this input* — never about the algorithm itself.

It is like judging which delivery rider is faster by timing one
delivery each. One rider hit traffic, another had a road closed —
the stopwatch tells us about the roads, not the riders. A better
comparison: one route has 10 stops, the other has 100. That holds
true in any traffic. The next sections build exactly this kind of
comparison for code.
