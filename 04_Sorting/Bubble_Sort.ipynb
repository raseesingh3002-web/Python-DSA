# BUBBLE SORT - Building the Algorithm Step by Step

**What is Bubble Sort?**  
Bubble sort sorts an array by repeatedly comparing adjacent numbers and if the adjacent numbers are in the wrong order bubble sort swaps them.  

With each iteration of comparison , the largest remaining element settles to it's correct position,i.e., at the end of the list.  

**Pass**  
one complete go through of the list from beginning to end, during which every adjacent pair of elements is compared once and swapped if found out of order.  





**How does a single pass efect the sorting in bubble sort?** 

```python
''' Effect of 1st pass:'''
def sorted_array(lst):
    n=len(lst)
    #outer loop represents the number of pass:
    for i in range(0,1):
        #inner loop represents the number of time we are comparing adjacent elements
        for j in range(0,n-1):
            #checking adjacent elements
            if lst[j]>lst[j+1]:
                #swapping
                a=lst[j]
                lst[j]=lst[j+1]
                lst[j+1]=a
    return lst

arr=[4,3,2,1]
sorted_array(arr)
```

As We can observe above, after 1st pass largest element has found it's correct place/order/index.

```python
''' Effect of 2nd pass:'''
def sorted_array(lst):
    n=len(lst)
    #outer loop represents the number of pass:
    for i in range(0,2):
        #inner loop represents the number of time we are comparing adjacent elements
        for j in range(0,n-1-i):
            #checking adjacent elements
            if lst[j]>lst[j+1]:
                #swapping
                a=lst[j]
                lst[j]=lst[j+1]
                lst[j+1]=a
    return lst

arr=[4,3,2,1]
sorted_array(arr)
```

After 2nd pass , the second largest element has found it's correct place/order/index.  
Here notice one thing that in the range of inner loop it is (n-1-i) this is done becasue after 1st pass the largest element has find it's correct place/order/index so there is no need to check the largest element with it's adjacent element again. This will prevent us doing the checking of those elements which are already placed at their correct place/order/index.

```python
''' Effect of 3rd pass:'''
def sorted_array(lst):
    n=len(lst)
    #outer loop represents the number of pass:
    for i in range(0,3):
        #inner loop represents the number of time we are comparing adjacent elements
        for j in range(0,n-1-i):
            #checking adjacent elements
            if lst[j]>lst[j+1]:
                #swapping
                a=lst[j]
                lst[j]=lst[j+1]
                lst[j+1]=a
    return lst

arr=[4,3,2,1]
sorted_array(arr)
```

We finally have found an sorted array after 3 passes i.e. after 'len(lst)-1' passes.

```python
'''FINAL BUBBLE SORT ALGORITHM'''
def bubble_sort(lst):
    n=len(lst)#length of list/array
    #outer loop which represents number of passes(should be n-1 for bubble sort)
    for i in range(0,n-1):
        #inner loop represents number of times we are comparing adjacent elements
        for j in range(0,n-1-i):
            #comparison of adjacent elements
            if lst[j]>lst[j+1]:
                #swapping
                a=lst[j]
                lst[j]=lst[j+1]
                lst[j+1]=a
    return lst
arr=[12,25,11,34,90,22]
bubble_sort(arr)
```
