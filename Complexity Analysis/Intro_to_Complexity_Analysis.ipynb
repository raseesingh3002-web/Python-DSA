# COMPLEXITY ANALYSIS

Any problem can be solved with different approaches. Complexity Analysis is the way to compare the different approaches that are used to solve a problem and decide which one is better.  

One Obvious way to compare two programs would be to run them and time them but fail for the following reason:  
- **same code can run faster on a new machine than an old one** (because new machine with relatively latest CPU than the old machine completes more instructions per second, Newer RAM is quicker and many other reasons too.)   

-  **faster in one programming language than other, ex:- faster in C than in Python**(C is translated into machine code once before running, so the CPU executes it directly; Python is translated line - by - line while running, adding overhead at every step - so the same logic runs slower)  

- **faster when the machine is not busy**(The CPU  is shared among all running programs, so the same code gives different timings depending on what else is running at that moment.)

So instead of asking "how fast does it run?" , Complexity Analysis asks a **<u>machine independent question: "how does the amount of work grow as the input grows?"</u>**.  
An algorithm that handles 100 orders easily might collapse at 10 lakh orders - growth behavior is what separates a solution that scales from one that doesn't.  

Every algorithm consumes two resources:  
- **Time** -> How many operations it performs(time complexity)  
- **Memory** -> how much extra space it needs(space complexity)  

Generally there's a trade off between that an individual can make code faster by using more memory, or save memory by doing more work. Complexity Analysis gives us the language to reason about that trade off and pick deliberately.
