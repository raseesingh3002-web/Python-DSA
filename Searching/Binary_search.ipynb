```python
def binary_search(array,target):
    size=len(array)-1
    start=0
    end=size
    while(start<=end):
        mid=(start+end)//2

        if(array[mid]==target):
            return mid
        elif (array[mid]>target):
            end=mid-1
        else:
            start=mid+1
    return -1
```

```python
my_list=[1,2,3,4,5]
target=3
binary_search(my_list,target)
```

```python
target=5
binary_search(my_list,target)
```

```python
target=9
binary_search(my_list,target)
```
